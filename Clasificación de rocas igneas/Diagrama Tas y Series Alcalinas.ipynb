```python
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
```

```python
filepath = Path('PracticaMagm√°ticos.xlsx')
df = pd.read_excel(filepath)
```

```python
df= df.transpose()
df.columns = df.iloc[0]
df = df[1:]
```

```python
#df 
```

```python
#Se establecen las reglas
NK= df["Na2O"]+df["K2O"]
SIO=df["SiO2"]
FM=df["FeO"]/df["MgO"]
```

```python
from pyrolite.plot.templates import TAS, SpinelFeBivariate
from pyrolite.util.plot.axes import share_axes

ax = TAS(linewidth=0.5, add_labels=True);plt.title("Diagrama TAS")
plt.scatter(SIO,NK)
for i, row in df.iterrows():
    plt.annotate(i, (SIO[i], NK[i]))
plt.show()
```

```python
plt.scatter(SIO,FM); plt.title('Serie Magm√°tica (Miyashiro,1974)')
y= 0.1562*SIO - 6.6838
plt.plot(SIO,y)
plt.xlim(45, 80)

quadrant_labels = {
    "Series Tole√≠ticas": (60, 6),
    "Series Calcoalcalinas": (60, -5),
}
for label, (x, y) in quadrant_labels.items():
    plt.text(x, y, label, ha='center', va='center')

for i, row in df.iterrows():
    plt.annotate(i, (SIO[i], FM[i]))
    
plt.ylabel('FeO/MgO');plt.xlabel('SIO2')

 
```
