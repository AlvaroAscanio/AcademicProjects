```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Dato = "/content/Datos_Roca.csv";
df= pd.DataFrame(pd.read_csv(Dato, delimiter=","))
df= df.transpose()
df.columns = df.iloc[0]
df = df[1:]

#Revisar que los datos hayan sido cargados correctamente.
```

```python
#Se establecen las reglas
ANK= df["Al2O3"]/(df["Na2O"]+df["K2O"])
ACNK=df["Al2O3"]/(df["CaO"]+df["Na2O"]+df["K2O"])
```

```python
plt.scatter(ACNK,ANK); plt.title('Clasificación de Muestras según el porcentaje de Alumina')
plt.xlabel('Al2O3/CaO+NaO+K2O');plt.ylabel('Al2O3/Na2O+K2O')
plt.axvline(x=1, color='red', linestyle='--')
plt.axhline(y=1, color='red', linestyle='--')
for i, row in df.iterrows():
    plt.annotate(i, (ACNK[i], ANK[i]))

    quadrant_labels = {
    "Peraluminosa": (1.5, 3),
    "Metaluminosa": (0.25, 3),
    "Peralcalina": (0.25, 0.5)
}
for label, (x, y) in quadrant_labels.items():
    plt.text(x, y, label, ha='center', va='center')


plt.show()
```
